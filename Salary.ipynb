### Анализ зависимости зарплаты и опыта работы

Данные были взяты с сайта `https://www.kaggle.com/datasets/vikrant9860/salary-csv/data`.
В данных два столбца: 
1. `YearsExperience` опыт работы в годах
2. `Salary` зарплата в у.е.

```python
#импортируем библиотеки
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#читаем и сохраем датасет в переменную
df = pd.read_csv('Salary.csv')
```

```python
#выводим первые строки фрейма
df.head()
```

```python
#смотрим на общую информацию
df.info()
```

```python
df.duplicated().sum()
```

**В данных нет пропусков и явных дубликатов. Типы данных соответствуют значениям.**

```python
#узнаем max, min, mean значения методом describe
df['YearsExperience'].describe()
```

```python
#строим гистограмму
df['YearsExperience'].plot(kind='hist', bins=10, title='Расспределение по опыту работы', figsize=(10,5), grid=True)
plt.show()
```

```python
#узнаем max, min, mean значения методом describe
df['Salary'].describe()
```

```python
#строим гистограмму
df['Salary'].plot(kind='hist', bins=10, title='Расспределение по зарплате', figsize=(10,5), grid=True)
plt.show()
```

**Max, min, mean значения для столбца "YearsExperience" составляют 1.1, 13.5, 6.3; для столбца "Salary" - 37 731, 139 465, 83 945. Из графика видно, что наибольшее количество сотрудников имееют опыт от 3 до 5 лет, и зарплату от 50 000 - 70 000 у.е.**

```python
#посмотрим на стаж работника с самой высокой зп

max_salary = df[df['Salary'] == 139465]
max_salary
```

**Работник с максимальной зарплатой имеет максимальный опыт работы**

```python
print(f"Корреляция между зарплатой и опытом работы составляет: {df['YearsExperience'].corr(df['Salary'])}")
```

```python
df.plot(x='Salary', y='YearsExperience', kind='scatter', grid=True, figsize=(10,5), alpha=1, title='График зависимости')
plt.show()
```

**Между опытом и зароботной платой сотрудника существует высокая корреляция: 0.98**
